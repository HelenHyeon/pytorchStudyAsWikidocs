# One-Hot Encoding

- **원-핫 인코딩** : 선택지의 개수만큼 차원을 가지며, 선택지에 해당하는 인덱스 원소에는 1, 나머지 원소는 0 값을 넣는 표현 방법이다.
- Ex) 강아지, 고양이, 냉장고라는 선택지에 각 0, 1, 2번 인덱스를 부여한다. 총 선택지는 3개이므로 모두 3차원 벡터이며 자신에게 해당하는 인덱스에만 1의 값을 가지고 있다.
```
강아지 = [1, 0, 0] 
고양이 = [0, 1, 0]
냉장고 = [0, 0, 1]
```

# 원-핫 벡터의 무작위성

- 각 클래스 간 관계가 균등할때 실제값을 원-핫 벡터는 다중 클래스 분류 문제에 대한 적절한 표현 방법이다.
- 다중 클래스 분류 문제에서는 클래스의 개수만큼의 숫자 레이블이 필요하다.
- Ex) 레이블 {Banana, Tomato, Apple} 을 분류할 때
    - 직관적 방법 : 정수 인코딩을 하여 각각 0, 1, 2로 레이블링 한다.
    - 하지만 위의 정수 인코딩보다 웟-핫 인코딩이 더욱 효과적인 표현 방법이다.

## 정수 인코딩 경우

- {Banana, Tomato, Apple}에 정수 인코딩을 사용하여 레이블에 각각 1, 2, 3을 부여할 경우
- 선형 회귀에서 배운 MSE(평균 제곱 오차)를 사용하면 문제가 나타난다.
![수식](../img/OHE1.png)
- 평균을 구하지 않고 제곱 오차로만 판단했을 때
    - 실제값이 Tomato일때 예측값이 Banana인 경우 : **(2 - 1)² = 1**
    - 실제값이 Apple일때 예측값이 Banana인 경우 : **(3 - 1)² = 4**
- 즉, 정수 인코딩은 순서에 따라 결과가 다르게 나타나기 때문에 **순서의 의미가 없는 클래스에서 사용하기 부적절하다.**

## 원-핫 인코딩의 경우

- {Banana, Tomato, Apple}에 원-핫 인코딩을 사용하여 각각 (1,0,0), (0,1,0), (0,0,1)을 부여할 경우
- 제곱 오차를 계산할 경우 정수 인코딩과 다르게 균등한 결과가 나온다.
![수식](../img/OHE2.png)
- 원-핫 벡터는 모든 쌍에 대해 유클리드 거리를 구할 경우 전부 동일하다.
- 이는 원-핫 벡터는 각 클래스의 표현 방법이 무작위성을 가진다는 것을 표현한다.
- 하지만 단어의 유사성을 구하는 등의 문제에서는 단점이라고 할 수 있다.
